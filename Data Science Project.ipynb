## Task 22: Take a sample of size 50 from the column damageDealt for 100 times and calculate its mean. Plot it on a histogram and comment on its distribution.

```python
mean = []
for i in range(100):
    mean.append(df.damageDealt.sample(50).mean())
mean_arr = np.array(mean)
mean_arr.mean()
```

```python
plt.hist(mean)
plt.xlabel('Damage Dealt')
```

Most times the damage dealt in the sample taken is between 120 & 130.
Number of times that damage dealt around 180 is less than 3, also the same goes for damage dealt being less than 100
