### Importing Libraries

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

## Question 1

plot a line plot between a and b:

    a=np.arange(40,50)
    b=np.arange(50,60) 

```python
a = np.arange(40,50)
b = np.arange(50,60)

plt.plot(a, b, marker = "*", color = "#07F")
plt.title("Line Plot", color = "#07F")
plt.xlabel("X-axis")
plt.ylabel("Y-Label")
```

## Question 2

Plot a line plot showing the sales trend in company 1 and 2:

    days = [1,2,3,4,5,6,7] #days of d week
    sales_1 = [160,150,140,145,175,165,180] #sales of company1
    sales_2 = [70,90,160,150,140,145,175]  #sales of company2

```python
days = range(1, 8)
sales_1 = [160,150,140,145,175,165,180]
sales_2 = [70,90,160,150,140,145,175]

plt.plot(days, sales_1, color = '#C00', marker = 'o', label = "Company 1")
plt.plot(days, sales_2, color = '#07F', marker = 's', label = "Company 2")
plt.title("Sales Trend", color = "#070")
plt.xlabel("Days", color = "#070")
plt.ylabel("Sales", color = "#070")
plt.legend()
```
