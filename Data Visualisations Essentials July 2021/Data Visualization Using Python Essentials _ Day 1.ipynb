```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sb
```

```python
x = np.arange(50)
y = np.random.randint(0, 100, 50)
print(x)
print(y)
```

```python
plt.plot(x, y, color = "#07F", marker = '+')
plt.title("Random Graph")
plt.xlabel("Days")
plt.ylabel("Count")
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)

plt.plot(x, y, color = "Red")
plt.title("Sin Wave")
```

## Assignment

```python
x = np.arange(10)
y = x * 3 + 10

plt.plot(x, y, 'red', marker = '*')
plt.title("3x + 10", color = 'blue')
plt.xlabel("x", color = 'blue')
plt.ylabel("y", color = 'blue')
```

```python
x = np.arange(0, 10)
y = x ** 2

plt.plot(x, y, color = '#050', marker = ".")
plt.title("Square Function")
```

```python
x = np.arange(0, 10)

plt.subplot(2, 2, 1)
plt.plot(x, x ** 2)
plt.subplot(2, 2, 2)
plt.plot(x, x ** 3)
plt.subplot(2, 2, 3)
plt.plot(x, x ** 4)
plt.subplot(2, 2, 4)
plt.plot(x, x ** 5)
```
