## Question 1
Plot using sns.relplot, line plot, box plot on fmri data set

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
fmri.shape
```

```python
fmri.info()
```

```python
fmri.describe()
```

### relplot

```python
sns.relplot(data = fmri, x = "timepoint", y = "signal", hue = "region", style = "event")
```

### line plot

```python
sns.relplot(data = fmri, x = "timepoint", y = "signal", kind = "line", color = "purple")
```

### box plot

```python
sns.catplot(data = fmri, x = "event", y = "signal", kind = "box")
```
