## Answer 2

#### Function to read lists

```python
def readList():
    lst = []
    for i in range(5):
        lst.append(input())
    return lst
```

### Converting lists to numpy arrays

```python
lst1 = readList()
lst2 = readList()

arr1 = np.array(lst1)
arr2 = np.array(lst2)

print(arr1)
print(arr2)
```

### Concatenating these arrays

```python
arr = np.concatenate((arr1,arr2))
arr
```

### Sorting the new array

```python
arr.sort()
arr
```
